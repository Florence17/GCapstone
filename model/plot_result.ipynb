```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.cm as cm
%matplotlib inline
sns.set()
```

```python

#Normalize confusion matrix:
def normalize(cm):
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    return cm



```

```python

def plot_confusion_matrix(confusion):
#     (bad: 1,2,3 stars; good: 4 star; best: 5 star)
    labels = ['bad', 'good', 'best']  
#     fig = plt.figure()
#     ax = fig.add_subplot(111)
    sns.heatmap(confusion, cmap="YlGnBu")
    plt.set_xticklabels([''] + labels)
    plt.set_yticklabels([''] + labels)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix of Random Forest Classifier')
    plt.savefig('confusion_final.png')
    plt.show()
```

```python
confusion_final = np.array([[66,  9,  1],
 [20, 45, 11],
 [16, 14, 46]])
```

```python
normalized_cm = normalize(confusion_final)
```

```python
plot_confusion_matrix(normalized_cm)
sns.heatmap(normalized_cm)
```
